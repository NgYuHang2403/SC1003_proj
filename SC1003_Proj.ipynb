```python
def TotalNum(v): #len
    #set amt variable to 0
    amt = 0
    #use for loop to count every item in the list v
    for i in v:
        amt += 1
    return amt
```

```python
def PList(list):
    for i in range(TotalNum(list)):
        print(str(list[i])+'\n')
```

```python
def Splitting_Group(array):
    #initiate empty variable SGroup and smallGroup
    SGroup = []
    smallGroup = []
    #initiate counter variable group
    group = 1
    #sort the data by Tutorial Group
    merge_sort(array,0,TotalNum(array)-1,'Tutorial Group')
    #start for loop
    for dict in array:
        #find the same tutorial group and add it into smallGroup list
        if dict['Tutorial Group'][2:] == str(group):
            smallGroup.append(dict)
        #increase group number when smallGroup list reach 50
        if TotalNum(smallGroup) == 50:
            group += 1
            #add smallGroup into SGroup
            SGroup.append(smallGroup)
            #reset smallGroup
            smallGroup = []
    return SGroup
```

```python
def readFile(fileName):
    with open(fileName) as file:
        column_names = file.readline()
        keys = column_names.split(",")
        keys[-1] = keys[-1].strip()
        number_of_columns = len(keys)
        list_of_dictionaries = []
        data = file.readlines()
        list_of_rows = []
        for row in data:
            list_of_rows.append(row.split(","))
            for a in list_of_rows:
                a[-1] = a[-1].strip()
        file.close()
        for item in list_of_rows:
            row_as_a_dictionary = {}
            for i in range(number_of_columns):
                row_as_a_dictionary[keys[i]] = item[i]
            list_of_dictionaries.append(row_as_a_dictionary)

    return list_of_dictionaries
```

```python
def writeFile(data):
    try:
        # Specify the filename 
        filename = 'NewRecord.csv' 
         
        # Open the file for writing 
        with open(filename, 'w') as file: 
            # Write the header 
            header = ','.join(data[0].keys())  # Get the keys from the first dictionary 
            file.write(header + '\n')  # Write header followed by a newline 
             
            # Write the data 
            for entry in data: 
                row = ','.join(str(entry[key]) for key in entry)  # Convert each value to string and join with commas 
                file.write(row + '\n')  # Write each row followed by a newline 
         
        print(f"Data exported to {filename}") 
    except:
        print(f'failed to export {filename}')
```

```python
def merge(arr, left, mid, right,sortBy):
    n1 = mid - left + 1
    n2 = right - mid

    # Create temp arrays
    L = []
    R = []

    # Copy data to temp arrays L[] and R[]
    for i in range(n1):
        L.append(arr[left + i])
    for j in range(n2):
        R.append(arr[mid + 1 + j])

    i = 0  # Initial index of first subarray
    j = 0  # Initial index of second subarray
    k = left  # Initial index of merged subarray
    # Merge the temp arrays back
    # into arr[left..right]
    if sortBy == 'Tutorial Group':
        while i < n1 and j < n2:
            if int(L[i][sortBy][2:]) <= int(R[j][sortBy][2:]):
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
    elif not sortBy:
        while i < n1 and j < n2:
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
    else:
        while i < n1 and j < n2:
            if float(L[i][sortBy]) <= float(R[j][sortBy]):
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

    # Copy the remaining elements of L[],
    # if there are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    # Copy the remaining elements of R[], 
    # if there are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def merge_sort(arr, left, right,sortBy):
    if left < right:
        mid = (left + right) // 2

        merge_sort(arr, left, mid,sortBy)
        merge_sort(arr, mid + 1, right,sortBy)
        merge(arr, left, mid, right,sortBy)
```

```python
def findDiff(data):
    lowgpa = []
    midgpa = []
    highgpa = []
    merge_sort(data,0,TotalNum(data)-1,'CGPA')
    lowestgpa = float(data[0]['CGPA'])
    highestgpa = float(data[TotalNum(data)-1]['CGPA'])
    gpaDiff = highestgpa - lowestgpa
    divide = gpaDiff/3
    return divide,lowestgpa
```
